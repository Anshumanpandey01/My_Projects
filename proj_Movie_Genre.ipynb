##  Extracting Top Movie Genre Per Year

Now we filter the movie dataset for the years we have crime data:

- 2006  
- 2008  
- 2013  
- 2019  
- 2020  
- 2021  
- 2022  

For each year, we will find the **most common genre** (main_genre) using `value_counts()`.

This will be used later to compare movie genres with crime rates.


```python
years = [2006, 2008, 2013, 2019, 2020, 2021, 2022]

top_genres = {}

for y in years:
    year_movies = movies[movies['Year'] == y]
    if not year_movies.empty:
        top_genre = year_movies['main_genre'].value_counts().idxmax()
        top_genres[y] = top_genre
    else:
        top_genres[y] = None  # if no movies for that year

top_genres
```

## Analysis: Impact of Dominant Movie Genre on Crime (per-year and overall)

This section creates figures to inspect whether the dominant movie genre in a given year (Top_Genre) co-occurs with higher or lower counts of selected violent crimes.

Figures included:
- Per-year bar charts of selected crimes (each plot shows the year's Top_Genre in the title)
- Bar charts comparing mean crime counts grouped by genre
- Scatterplots: genre code vs each crime type
- Correlation heatmap including Genre_Code


```python
# Per-year bar charts: for each year plot the 4 selected crimes and show Top_Genre in the title
import matplotlib.pyplot as plt

crime_cols = ['Murder','Attempt_Murder','Rape','Kidnapping']
years_available = crime_df_final['Year'].tolist()

for idx, row in crime_df_final.iterrows():
    year = int(row['Year'])
    vals = [row.get(c, None) for c in crime_cols]
    # skip if all None
    if all(pd.isna(v) for v in vals):
        continue
    plt.figure(figsize=(6,4))
    bars = plt.bar(crime_cols, [0 if pd.isna(v) else v for v in vals])
    plt.title(f'Crime counts in {year} (Top genre: {row["Top_Genre"]})')
    plt.ylabel('Number of cases')
    for b, v in zip(bars, vals):
        if not pd.isna(v):
            plt.text(b.get_x() + b.get_width()/2, v + max(1, 0.01*max([x if (not pd.isna(x)) else 0 for x in vals])), f'{int(v):,}', ha='center', va='bottom', fontsize=9)
    plt.ylim(0, max([0 if pd.isna(x) else x for x in vals])*1.15)
    plt.tight_layout()
    plt.show()
```

### Per-year crime rates by genre (population-normalized)
This shows crime rates (per 100,000) for each year with the top genre in the title for context. Population is mapped approximately — adjust if you have exact population data.


```python
# approximate population mapping (same as earlier)
population = {2006:1150000000,2008:1180000000,2013:1250000000,2019:1360000000,2020:1380000000,2021:1390000000,2022:1400000000}
crime_df_final['Population'] = crime_df_final['Year'].map(population)
# safe rate calculation
for c in ['Murder','Rape','Kidnapping','Attempt_Murder']:
    rate_col = c + '_rate100k'
    crime_df_final[rate_col] = crime_df_final.apply(lambda r: (r[c]/r['Population']*100000) if pd.notna(r[c]) and pd.notna(r['Population']) else None, axis=1)

# plot rates per year with genre in title
for idx, r in crime_df_final.iterrows():
    y = int(r['Year'])
    rates = [r.get(c+'_rate100k', None) for c in ['Murder','Attempt_Murder','Rape','Kidnapping']]
    if all(pd.isna(v) for v in rates): continue
    plt.figure(figsize=(6,4))
    bars = plt.bar(['Murder','Attempt_Murder','Rape','Kidnapping'], [0 if pd.isna(v) else v for v in rates])
    plt.title(f'Crime rates per 100k in {y} (Top genre: {r["Top_Genre"]})')
    plt.ylabel('Rate per 100,000')
    for b, v in zip(bars, rates):
        if not pd.isna(v): plt.text(b.get_x()+b.get_width()/2, v + 0.0001, f'{v:.2f}', ha='center', va='bottom', fontsize=9)
    plt.tight_layout()
    plt.show()
```

## Regression Analysis  
### Does Movie Genre + Year Predict Crime?

To understand whether movie genre influences crime trends, we build a simple **Linear Regression model**:

**Target (y):**
- Murder Cases (you can change to Rape, Kidnapping, etc.)

**Features (X):**
- Year  
- Genre_Code (numeric encoding of Action/Comedy)

This helps us check:
- If Genre_Code has a positive or negative coefficient  
- If Year has increasing or decreasing trend  
- How well the model fits crime data (R² score)

If Genre_Code coefficient is close to **0**, then genre has **no measurable impact** on crime.


```python
from sklearn.linear_model import LinearRegression

# Define features and target
X = crime_df_final[['Year','Genre_Code']]
y = crime_df_final['Murder']

# Build and train model
model = LinearRegression()
model.fit(X, y)

# Print results
print("Regression Results")
print("-------")
print("R² Score:", model.score(X, y))
print("Coefficients:", dict(zip(X.columns, model.coef_)))
print("Intercept:", model.intercept_)
```

### Regression Interpretation

The regression model produced an R² score of 0.72, meaning 72% of murder variation is explained by Year + Genre_Code.

However:

- The coefficient for **Year** is +1626, showing crime increases over time.
- The coefficient for **Genre_Code (-6470)** incorrectly suggests Comedy reduces crime.

This is **not a real effect**, because Comedy appears only in **one year (2006)** and all remaining years are Action.  
Therefore, Genre_Code is artificially correlated with Year, creating a **false negative** effect.

 **Conclusion:**  
Movie genre does **not** meaningfully predict crime.  
The increase in crime is driven by **Year/population growth**, not movie content.


## Regression Without 2013 (Outlier Removal)

2013 crime numbers are extremely large due to different NCRB reporting format.  
We repeat the regression after removing 2013 to see if results change.


```python
clean = crime_df_final[crime_df_final['Year'] != 2013].copy()

X_clean = clean[['Year','Genre_Code']]
y_clean = clean['Murder']

model_clean = LinearRegression()
model_clean.fit(X_clean, y_clean)

print("Regression Results (2013 removed)")
print("R² Score:", model_clean.score(X_clean, y_clean))
print("Coefficients:", dict(zip(X_clean.columns, model_clean.coef_)))
print("Intercept:", model_clean.intercept_)
```

## Regression Interpretation (2013 Removed)

After removing the outlier year (2013) and rerunning the regression:

- R² = 0.97 (excellent model fit)
- Year coefficient = +2207 → crime rises over time
- Genre coefficient = +3514 → looks large, but **NOT meaningful**

### Why Genre is NOT meaningful:
- Only one Comedy year exists (2006)
- 2006 naturally has low crime because it is an early year
- The model mistakenly connects "Comedy" with low crime
- This is a false effect caused by **multicollinearity** (Genre depends on Year)

### Final Answer:
Movie genre changes do **not** affect crime trends.  
Crime increases mainly due to **timeline (Year)**, not movie content.


## Final Conclusion: Does Movie Genre Impact Crime?

After analyzing crime data across multiple years (2006–2022) and comparing it with movie genre trends:

### Key Findings

### Final Judgment
